#### [문제05]
#### 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
         
   * 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
          
         >>> addall([1])
            1
          >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
            45  
            
       
   * 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
      
          >>> addallodd([1])
            1
            >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
            25
   * [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)  
   * [참고 2]: sum() 등의 내장함수 사용 가능함

```python
def addall(a):
    sum=0
    for i in a:
        sum+=i
    return sum

def addall2(a):
    return sum(a)
#(sum(a) for i in a)
def addallodd(a):
    sum = 0
    for i in a: 
        if i%2==1:
            sum+=i
    return sum

def addallodd2(a):
    return sum(i for i in a if i % 2 == 1)
    
    
a = [1,2,6,4,5]
res = addall(a)
res2 = addall2(a)
print(res)
print(res2)
oddres = addallodd(a)
oddres2 = addallodd2(a)
print(oddres)
print(oddres2)
```

addall은 리스트의 원소개수만큼 반복하여 sum은 list원소들의 모든 합을 반환한다

addall2는 sum내장함수를 이용하여 list원소들의 모든 합을 반환한다

addallodd는 홀수의 조건인 나누어서 나머지가 1이면 sum에 더하여 list의 홀수 원소들의 모든 합을 반환한다

addallodd2는 sum 내장함수를 이용하여 홀수의 조건인 나누어서 나머지가 1이면 sum에 더하여 list의 홀수 원소들의 모든 합을 반환한다

   * [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.
   * 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
   * 2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
   * 3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
   * 4) 값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
t=[('one',1), ('two',2), ('three',3), ('four',4), ('five',5)]
t.sort()
print(t)
t.sort(reverse=True)
print(t)
def cmp_num(a):
    return a[1]
t.sort(key=cmp_num)
print(t)
t.sort(key=cmp_num,reverse = True)
print(t)
```

#### [문제08]
#### [ACM-ICPC 문제 1673번]
                    
   * 치킨 쿠폰
      
      https://www.acmicpc.net/problem/1673



```python
import sys
for line in sys.stdin:
    s=line.split()
    n = int(s[0])
    k = int(s[1])
    coupon = n
    total = n
    while coupon >= k:
        total+=coupon//k
        a = coupon//k
        b = coupon%k
        coupon = a + b
    print(total)
```

입력받은 s를 띄어쓰기 단위로 나누어 n과k에 대입한다
for문은 입력이 안들어 오면 자동으로 멈춘다
쿠폰수가 치킨수랑 처음에 같으니 n을 대입한다
치킨 수는 coupon 나누기 k한 값 쿠폰은 쿠폰 나누기 k한 값과 쿠폰 나누기 k한 나머지의 합이다
