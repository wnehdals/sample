#### 2) 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오
   * 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
   * 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
   * 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
   * 다음과 같은 실행 및 출력 결과도 도출되어야 한다.
      * div_path('/usr/local/bin/python')
         - ['/usr/local/bin','python']
      * div_path('/home/chulsoo/text.txt')
         - ['/home/chulsoo','text.txt']
      * [참고] 리스트(ㅣ) 내에 새로운 정수값(예를 들어 10)을 넣는 방법은 ㅣ.append(10)이다.  

```python
def div_path(s):
    a = s.count('/')       #'/'의 개수
    t=0                    #s속에서 제일 끝에 있는'/'찾기위한 변수
    for i in range (len(s)):
        if(s[i]=='/' ):
            t+=1
        if(t == a-1):
            res1 = s[0:i+1] #s의 처음부터 제일 끝에 있는'/'전까지 문자 저장
            b = i+1         #s의 제일 끝에 있는'/'위치 저장
            
            
    res2 = [res1,s[b+1:]]    
    print(res2)
    
s =input()
div_path(s)
```

s.count를 사용하여 '/'의 개수를 알아낸 후 s를 len(s)로 s의 길이만큼 반복하여 '/'를 발견하면 t값을 증가시킨다.
t의 값이 '/'전체 개수-1이랑 같다면 그'/'앞뒤로 경로명과 파일명으로 나누어진다 그러므로 t의값이 a-1과 같을 때 res1에 s의처음부터 i+1까지를 저장한다. res2f를 리스트로 만들기 위해res2=[res1, s[b+1:]]를 해주었다.b는s의 제일 끝에 있는'/'의 위치임으로 파일명은 b+1부터 끝까지로 지정해주고 res2에 저장한다.

#### 3) 두 개의 리스트를 인자로 받아서 그 두개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
   * 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
   * 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의      모든 원소를 넣어 반환한다.
   * 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
   * 반환하는 리스트는 정렬이 되어 있어야 한다.
   * 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
       * list_union([1, 2, 3], [1, 2, 4])
            - [1, 2, 3, 4]
       * list_union([-10, -5, 0, -1], [100, 9, 0, 9])
            - [-10, -5, -1, 0, 9, 100]
       * list_union([0, 1, 2], [0, 1, 2])
            - [0, 1, 2]
   * [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append      (10) 이다.
   * [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in      l 이다.

```python
def list_union(lista, listb):
    for i in range (len(listb)):
         if(listb[i] not in lista):
            k = listb[i]
            lista.append(k)  
            
            
    for i in range (len(lista)):
        index = i
        for j in range(i+1,len(lista)):
            if(lista[index]>lista[j]):
                index = j
        
        
        temp = lista[index]
        lista[index] = lista[i]
        lista[i] = temp
            
            
    return lista

lista = [1,3,2]
listb = [1,2,4]
union = list_union(lista,listb)
print(union)
lista = [-10,-5,0,-1]
listb = [100,9,0,9]
union = list_union(lista,listb)
print(union)
lista = [0,1,2]
listb = [0,1,2]
union = list_union(lista,listb)
print(union)
```

합집합을 만들기 위해서는 lista에는 없고 listb에만 있는 원소들만을 lista에 넣어주면 된다. 그 후에 lista의 원소들을 작은순서부터 정렬시키면 된다.

```python
def ab(n,a):
    p =''
    for i in range(len(a)):
        res = a[i]
        p+= res*int(n)
        
    return p




r =[]
n=int(input())
for i in range(n):
    s=input()
    k=s.split()
    n1=k[0]
    t=k[1]
    r.append(ab(n1,t)) 
for i in range(n):
    print(r[i])



```

split함수를 사용하여 숫자와 문자를 공백으로 구분하였다. 숫자는 n1에 저장 문자는t에 저장한다. 함수 ab는 숫자와 문자를 받아 각 문자를 숫자만큼 반복하여 p에 저장한다. 그리고 p를 반환한다.리스트 r에 함수ab의값을 리스트안에 넣어 리스트의원소들을 출력한다.
