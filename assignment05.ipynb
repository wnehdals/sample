```python
class Car:
    def speed(self):
        print("50km/s")

class Porsche(Car):
    def speed(self):
        print("100km/s")
        
class Lamborghini(Car):
    def speed(self):
        print("120km/s")
        
class Ford(Car):
    def speed(self):
        print("150km/s")
        
for each in (Porsche(),Lamborghini(),Ford()):
    each.speed()
```

    다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

   * 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

      * count: Counter 인스턴스가 지니는 초기 정수 값
      * step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
        
            >>> c = Counter(10) 
            >>> d = Counter(10, 2) 

   * 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

             >>> print(c) 
            [Count (step: 1)] 10 
             >>> print(d) 
            [Count (step: 2)] 10 

   * 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

            >>> c.incr() 
            >>> d.incr() 
            >>> print(c) 
            [Count (step: 1)] 11 
            >>> print(d) 
            [Count (step: 2)] 12 

   * 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출 (call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

            >>> c() 
            >>> d() 
            >>> print(c) 
            [Count (step: 1)] 12 
            >>> print(d) 
            [Count (step: 2)] 14 

   * 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

            >>> c = c + 5 
            >>> d = d - 5 
            >>> print(c) 
            [Count (step: 1)] 17 
            >>> print(d) 
            [Count (step: 2)] 9 

   * 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

            >>> print(c) > 10 
            True 
            >>> print(d) > 10 
            False 
            >>> print(c) < 10 
            False 
            >>> print(d) < 10 
            True 
            >>> print(c) == 17 
            True 
            >>> print(d) != 9 
            False 

```python
class Counter:
    def __init__(self,number,step=1):#생성자
        self.number=number
        self.step=step
    def __str__(self):
        return "[Count(step:"+str(self.step)+")] "+str(self.number)
    def incr(self):
        self.number+=self.step
    def __call__(self):              #호출함수
        self.number+=self.step
    def __add__(self,other):        #+연산자 중복 정의
        self.number = self.number + other
        return self
    def __sub__(self,other):        #-연산자 중복 정의
        self.number = self.number - other
        return self
    def __gt__(self,y):             #>연산자 중복 정의
        if isinstance(y, Counter):  #객체끼리 비교
            return self.number > y.number
        elif isinstance(y, int):    #객체와 정수비교
            return self.number > y
        else:
            return False
    def __lt__(self, y):           #<연산자 중복정의
        if isinstance(y, Counter): #객체끼리 비교
            return self.number < y.number
        elif isinstance(y, int):   #객체와 정수비교
            return self.number < y
        else:
            return False
    def __eq__(self, y):           #==연산자 중복정의
        if isinstance(y, Counter): #객체끼리 비교
            return self.number == y.number
        elif isinstance(y, int):   #객체와 정수비교
            return self.number == y
        else:
            return False    
    def __ne__(self, y):         #!=연산자 중복정의
        if isinstance(y, Counter):#객체끼리 비교
            return self.number != y.number
        elif isinstance(y, int):   #객체와 정수 비교
            return self.number != y
        else:
            return False
c=Counter(10)
d=Counter(10,2)
c.incr()
d.incr()
print(c)
print(d)
c()
d()
print(c)
print(d)
c=c+5
d=d-5
print(c)
print(d)
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

[4번 문제]
 
     다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

    * MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

* elimicate_duplicate()

        elimicate_duplicate(l)함수는 static메쏘드로 만들어졌다. static메쏘드는 클래스이름을 이용하여 직접 호출이 가능하다
        이 함수는 s라는 빈 리스트를 만들고 매개변수로 ㅣ을 받아 ㅣ이 s의 리스트에 있는 값과 ㅣ값이 같지 않다면 리스트에 그 변수를 추가하여 리스트s에 중복된 값이 없도록 한다
        l[:] = []하여 리스트l을 빈 상태로 만든 후 리스트s의 원소들을 l에 추가한다
* __init__()
    
        __init__()은 생성자라고 한다 생성자란 클래스가 불리면 제일 먼저 호출되는 함수이다.매개변수로 self를 가져야 한다.
        __init__(self,l)은 MySet의 생성자로써 MySet classs는 list임으로 리스트에 값을 추가하는 메소드인 append(e)를 사용하여 e를 리스트의 원소로 추가한다. 
        위의 eliminate_duplicate(l)에서 말했듯이 static매쏘드로써 클래스이름을 이용하여 원소의 중복을 처리하는 eliminate_duplicate함수를 호출하여 중복이 제거된 리스트를 생성한다.
* __str__()
          
        __str__()은 print함수를 하였을 떄 출력되는 내용을 설정하는 함수로 먼저 "MySet : {" 다음 원소들이 들어가야 하니 for문을 사용하여 MySet객체의 원소들을 str(e)를 하여 문자로 추가하고 
        " ,"를 붙여"MySet : {1 ,2 , 3, ,"의 셍성되고 3뒤에 " ,"을 제거하기위해 슬라이싱을 활용하여 result[0:len(result)-2]을 하게되면 0부터 result의길이-1에 해당하는인덱스의 문자열만 result에 저장한다.
        그 이후 괄호를 닫는"}"를 추가하여 {1 ,2 ,3}이란 형태의 result변수를 반환하게되는 함수이다.

[5번 문제]

    4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

    * 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
    * 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

    >>> u = s | t 
    >>> print(u) 
    MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 

    * 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

    >>> u = s & t 
    >>> print(u) 
    MySet: {2, 3} 

    * 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    >>> s = MySet([1, 2, 3]) 
    >>> t = MySet([3, 4, 5]) 
    >>> u = s - t 
    >>> print(u) 
    MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    def __or__(self,y):#|연산자를 중복정의 할 떄 __or__메소드를 사용한다
        self=self+y     #self와y의 원소를 합한다
        MySet.eliminate_duplicate(self)#중복된 원소를 제거한다
        return MySet(self)
      
    def __and__(self,y):#연산자를 중복정의 할 때 __and__메소드를 사용한다
        s=[]
        for i in range(len(self)):#이중반복문을 사용하여 self와y의 원소가 같으면 리스트s에추가
            for j in range(len(y)):
                if self[i] == y[j]:
                    s.append(self[i])
        return MySet(s)
    def __sub__(self,y):#-연산자를 중복정의 할 때 __sub__메소드를 사용한다
        s=self&y        #self와y의 교집합을 s에 저장
        k=[]
        for i in range(len(self)):#self원소에서 s의원소와 같은게 있다면 리스트k에 저장하지 않는다
            for j in range(len(s)):
                if self[i] != s[j]:
                    k.append(self[i])
        return MySet(k)
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s | t 
    print(u) 
    u = s & t 
    print(u) 
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5]) 
    u = s - t 
    print(u) 

```
